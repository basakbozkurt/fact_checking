```python
# Extract the year data as an integer from the 'Year founded' column
years = df['Year founded'].dt.year

# Determine the range of years
min_year, max_year = years.min(), years.max()

# Create a histogram of the 'Year founded' column with bins for every year in the range
ax = years.plot(kind='hist', bins=range(min_year, max_year + 1), title='Year founded', align='left', rwidth=0.8)

# Customize the histogram appearance
ax.spines[['top', 'right']].set_visible(False)
ax.set_xticks(range(min_year, max_year + 1))
ax.set_xlabel('Year founded')

# Add count numbers on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom')

# Rotate the year labels vertically
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Convert date strings to datetime objects
df['Year founded'] = pd.to_datetime(df['Year founded'], format='%Y')
df['Last Post'] = pd.to_datetime(df['Last Post'], format='%m/%d/%Y')

# Create a new column 'Active Years' to calculate the duration for which each URL was active
df['Active Years'] = df.apply(lambda row: range(row['Year founded'].year, row['Last Post'].year + 1), axis=1)

# Flatten the list of active years
active_years = [year for sublist in df['Active Years'] for year in sublist]

# Count the occurrences of each year
year_counts = pd.Series(active_years).value_counts().sort_index()

# Plot the graph
plt.figure(figsize=(10, 6))
ax = year_counts.plot(kind='bar')
plt.title('Active Years of URLs')
plt.xlabel('Year')
plt.ylabel('Number of Active URLs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels on top of each bar
for i in ax.patches:
    ax.text(i.get_x() + i.get_width() / 2, i.get_height() + 0.1, str(int(i.get_height())), ha='center')

plt.tight_layout()
plt.show()

```

```python
# Flatten the list of active years and associate them with IFCN Code
active_years_data = []
for _, row in df.iterrows():
    membership_status = 'Member' if row['IFCN Code'] in ['Verified Signatory', 'Signatory'] else 'Non-Member'
    active_years_data.extend([(year, membership_status) for year in row['Active Years']])

# Create a DataFrame with the active years data
active_years_df = pd.DataFrame(active_years_data, columns=['Year', 'Membership Status'])

# Count the occurrences of each year per Membership Status
year_counts = active_years_df.groupby(['Year', 'Membership Status']).size().unstack(fill_value=0)

# Define colors for each membership status
membership_colors = {
    'Member': 'lightgreen',
    'Non-Member': 'green'
}

# Create a list of colors in the same order as the columns of the year_counts DataFrame
column_colors = [membership_colors.get(status, 'black') for status in year_counts.columns]

# Calculate the relative percentages
year_counts_percentages = year_counts.div(year_counts.sum(axis=1), axis=0) * 100

# Plot the stacked bar graph
ax = year_counts.plot(kind='bar', stacked=True, color=column_colors, figsize=(10, 6))

# Add percentage labels in black and vertical
for i, (year, row) in enumerate(year_counts.iterrows()):
    cumulative_height = 0
    for col in year_counts.columns:
        height = row[col]
        percentage = year_counts_percentages.loc[year, col]
        if height > 0:
            ax.text(
                i, cumulative_height + height / 2,
                f'{percentage:.1f}%',
                ha='center', va='center', fontsize=9, color='black', rotation=90
            )
        cumulative_height += height

# Add labels and a title to the plot
plt.title('Distribution of Fact-Checking Organisations Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Active Fact-Checking Organisations')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
